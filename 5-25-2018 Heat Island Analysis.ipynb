```python
import pandas as pd
```

```python
df1 = pd.read_csv("cit_015_urban_heat_island_effect.csv")
df1.head()
```

```python
df1.sort_values('d_t_diff', ascending = False, inplace = True)
highest = df1.head(10)
highest
#highest.to_csv('highest_diff_temp')
```

# Results
Out of the 10 cities that experience the highest urban heat island effect, 7 of them are in India (Jamkhed, Heggaddevankote, Mukhed, Haliyal, Namagiripettai, Titabortown, and Dabra). The others in the top 10 include Pedro Carbo, Ecuador, Kaunakakai, HI, USA, and Kajy-Say, Kyrgyzstan.

```python
#group by country
countries = df1.groupby('iso3').describe()
countries.d_t_diff
diff = countries.d_t_diff
diff.sort_values('mean', ascending = False)
```

```python
df1.set_index("name", inplace=True)
```

```python
df1.loc['Karachi']
```

```python
#Import WRI countries table
countries = pd.read_excel('countries.xls')
countries.head()
```

```python
#Join countries and heat island by countries to sort by regions and make understand ISO names
country_heat = pd.merge(diff, countries, how='inner', right_on= 'iso_a3', left_index= True)
```

```python
country_heat.sort_values('mean', ascending = False)
```

# Results

Malta experiences the highest heat island effect at 7.07 degrees C above the buffer followed by Bahrein and Barbados. However this mean may not indicate which countries have significant issues, because other cities experience this effect with temperature differences that are double these average, but because of the includion of many more cities in that country, they do not show up significantly when aggregating at the country level. 

```python
country_heat.groupby('wb_region').mean().sort_values('mean', ascending = False)
```

# Results
On average Latin American and the Carribbean experience the highest urban heat island effect at 1.115 deg. C above the surrounding buffer, followed by East Asia & the Pacific at ~1.0 deg C.
